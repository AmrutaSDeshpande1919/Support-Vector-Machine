```python
sal_train=pd.read_csv("C:\\Users\\HP\\Desktop\\ExcelR\\Assignment\\Assignment 14 SVM\\SalaryData_Train(1).csv")
sal_test=pd.read_csv("C:\\Users\\HP\\Desktop\\ExcelR\\Assignment\\Assignment 14 SVM\\SalaryData_Test(1).csv")
```

```python
string_col=['workclass','education','maritalstatus','occupation','relationship','race','sex','native']
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
for i in string_col:
    sal_train[i]=label_encoder.fit_transform(sal_train[i])
    sal_test[i]=label_encoder.fit_transform(sal_test[i])
```

```python
sal_train['Salary'] = label_encoder.fit_transform(sal_train['Salary'])
```

```python
train_x=sal_train.iloc[0:,0:13]
train_y=sal_train.iloc[0:,13]
test_x=sal_test.iloc[0:,0:13]
test_y=sal_test.iloc[0:,13]
```

```python
from sklearn.svm import SVC
```

```python
#kernel=rbf
model_rbf=SVC(kernel='rbf')
```

```python
model_rbf.fit(train_x,train_y)
```

```python
train_pred_rbf=model_rbf.predict(train_x)
```

```python
test_pred_rbf=model_rbf.predict(test_x)
```

```python
train_rbf_acc=np.mean(train_pred_rbf==train_y)
```

```python
test_rbf_acc=np.mean(test_pred_rbf==test_y)
```
